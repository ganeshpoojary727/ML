```python
import csv

data = list(csv.reader(open('enjoysport.csv')))
print("Training instances:\n", data[1:])

h = ['0'] * (len(data[0]) - 1)

for row in data[1:]:
    if row[-1] == 'yes':
        for j in range(len(h)):
            if h[j] == '0':
                h[j] = row[j]
            elif h[j] != row[j]:
                h[j] = '?'

print("Final hypothesis:", h)
```
