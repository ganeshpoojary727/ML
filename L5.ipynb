```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


np.random.seed(42)
X = np.random.rand(100).reshape(-1,1)


y = ['Class1' if v<=0.5 else 'Class2' for v in X[:50].ravel()]
true = ['Class1' if v<=0.5 else 'Class2' for v in X[50:].ravel()]


for k in [1,2,3,4,5,20,30]:
    pred = KNeighborsClassifier(k).fit(X[:50], y).predict(X[50:])
    
    print(f"\nk = {k}")
    print("Accuracy:", accuracy_score(true, pred)*100)
    print("Predictions:", pred)
```
